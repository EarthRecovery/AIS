```python
import json
```

```python
metadata_path = "./herta/meta.json"
history_data = ["./history/a1.json", "./history/a2.json", "./history/a3.json", "./history/a4.json", "./history/a5.json", "./history/a6.json"]
text_data = []
```

```python
metadata = json.load(open(metadata_path, "r", encoding="utf-8"))
text_data.append(metadata["herta"]["description"])
for data in metadata["herta"]["story"]:
    text_data.append(data)
for data in metadata["herta"]["video"]:
    text_data.append(data)

text_data.append(metadata["The herta"]["description"])
# text_data.append(metadata["The herta"]["story"])
for data in metadata["The herta"]["story"]:
    text_data.append(data)
for data in metadata["The herta"]["video"]:
    text_data.append(data)

for data in text_data:
    print(data)
    print("-----")
```

```python
for history_path in history_data:
    history = json.load(open(history_path, "r", encoding="utf-8"))
    for chapter_dialogue in history:
        for dialogue in chapter_dialogue["dialogues"]:
            for i in range(len(dialogue["dialogues"])):
                if "黑塔" in dialogue["dialogues"][i]:
                    text = dialogue["dialogues"][i]
                    if i != 0:
                        text += " " + dialogue["dialogues"][i-1]
                    if i != len(dialogue["dialogues"]) - 1:
                        text += " " + dialogue["dialogues"][i+1]
                    text_data.append(text)

        
```

```python
with open("output.json", "w", encoding="utf-8") as f:
    json.dump(
        text_data,
        f,
        ensure_ascii=False,  # 保留中文
        indent=2             # 美化缩进
    )
```
